```python

from geopy.geocoders import Nominatim

def get_coordinates(state_name):
    geolocator = Nominatim(user_agent="shopping_trends")  
    location = geolocator.geocode(state_name, language='en')
    if location:
        return location.latitude, location.longitude
    else:
        return None

df= pd.read_csv('Data/shopping_trends.csv')
state_column = 'Location'

df['Latitude'] = None
df['Longitude'] = None

for index, row in df.iterrows():
    state_name = row[state_column]
    coordinates = get_coordinates(state_name)
    if coordinates:
        df.at[index, 'Latitude'] = coordinates[0]
        df.at[index, 'Longitude'] = coordinates[1]
df.to_csv('Data/shopping_trends_geo.csv', index=False)

print(f"Latitude and Longitude added and DataFrame saved ")
```
